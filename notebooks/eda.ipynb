```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../processed/cleaned_sales_data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
total_sales = df['sales'].sum()
print(total_sales)
```

```python
total_profit = df['profit'].sum()
print(total_profit)
```

```python
avg_sales = df['sales'].mean()
print(avg_sales)
```

```python
monthly_sales = df.groupby('month_name')['sales'].sum()
print(monthly_sales)
```

```python
plt.figure(figsize=(12,6))
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
monthly_profit = df.groupby('month_name')['profit'].sum()
```

```python
plt.figure(figsize=(12,6))
monthly_profit.plot(kind='bar')
plt.title('Monthly Profit Analysis')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.show()
```

```python
top_products = df.groupby('product_name')['sales'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))
top_products.plot(kind='bar')
plt.title('Top 10 Selling Products')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
category_sales = df.groupby('category')['sales'].sum()
```

```python
plt.figure(figsize=(8,8))
category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Category-wise Sales Distribution')
plt.ylabel('')
plt.show()
```

```python
region_sales = df.groupby('region')['sales'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=region_sales.index, y=region_sales.values)
plt.title('Region-wise Sales')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['quantity'], bins=20)
plt.title('Quantity Distribution')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['sales'], y=df['profit'])
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
correlation = df.corr(numeric_only=True)
print(correlation)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
yearly_sales = df.groupby('year')['sales'].sum()
```

```python
plt.figure(figsize=(10,6))
yearly_sales.plot(kind='bar')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
top_customers = df.groupby('customer_name')['sales'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))
top_customers.plot(kind='bar')
plt.title('Top Customers')
plt.xlabel('Customer')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```
